```
import numpy as np
```

```
def OMC(n):
  sumvalue=0
  N=0
  for i in range(n):
    y=np.random.uniform(0,1)
    if y>0 and y<1/4:
      sumvalue+=2
      N+=1
    if y>1/4 and y<1:
      sumvalue+=2/3
      N+=1
  return sumvalue/N
```

```
print('expectation generated by OMC is ', OMC(10000000))
```

```
def OMC(n):
  sumvalue=0
  for i in range(n):
    b=np.random.normal(0,1)
    if b < -2:
      sumvalue+=1
  return sumvalue/n
```

```
print(' forward price generated by OMC is ', OMC(1000000))
```

```
from scipy.stats import norm
```

```
def IS(n,α):
  sumvalue=0
  for i in range(n):
    y=np.random.uniform(0,1)
    x=norm(-α,1).ppf(y)
    if x<-2:
      sumvalue+=np.exp(0.5*(α**2)+α*x)
  return sumvalue/n
```

```
print(' forward price generated by IS(2) is ', IS(1000000,2))
```
